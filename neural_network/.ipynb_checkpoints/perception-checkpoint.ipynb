```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
import os
```

```python
np.random.randn(num_samples, 2)
```

```python
# 生成 2D 分类数据
np.random.seed(42)
num_samples = 50
#randn是标准正态分布，数据点是围绕找原点的。现在的坐标轴是横轴x1，纵轴x2，正类就是将这些点移到第四象限，负类就是移到第二象限
X1 = np.random.randn(num_samples, 2) + np.array([2, -2])  # 正类 (+1)
X2 = np.random.randn(num_samples, 2) + np.array([-2, 2])  # 负类 (-1)
# print(X1)
X = np.vstack((X1, X2))#竖向拼接
print(X.shape)
y = np.hstack((np.ones(num_samples), -np.ones(num_samples)))  # 类别标签：+1 或 -1
print(y.shape)
```

```python
# 添加偏置项 x0 = 1  X （100，2） 100个样本，二维特征  然后加上 偏置项 1 （100，1），最后形成 （100，3）方便矩阵乘法
X = np.hstack((np.ones((X.shape[0], 1)), X))  # 变为 (1, x1, x2)
# print(X.shape)
# 初始化感知机参数 (w0, w1, w2)
w = np.random.randn(3)  # 随机初始化权重,因为这是单层感知机，w1是第一个特征的权重，w2是第二个特征的权重，w0会和1相乘，成为偏置项
# # 记录训练过程中的权重变化
history = [w.copy()]
# print(history)

# # 设置学习率
eta = 0.05
epochs = 10  # 训练轮数

# # 创建文件夹用于存储帧
frame_folder = "frames"
if not os.path.exists(frame_folder):
    os.makedirs(frame_folder)
```

```python
w.shape,X[0].shape
```

```python
np.dot(w,X[0])# w0,w1,w2 和 b,第一个特征，第二个特征  相乘累加求和
```

```python
# 感知机学习算法
for epoch in range(epochs):
    for i in range(X.shape[0]):#迭代每一个样本
        if np.sign(np.dot(w, X[i])) != y[i]:  # 如果误分类，则更新权重，sign是符号函数
            w += eta * y[i] * X[i]#求偏导后的，需要更新权重的值
            history.append(w.copy())  # 记录权重变化
```

```python
np.array([-4, 4]) * w[1]
```

```python
# 绘制并保存帧
frame_paths = []
print(history)
for frame, w in enumerate(history):
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # 画出数据点
    ax.scatter(X1[:, 0], X1[:, 1], color='b', label="Class +1")#正例（50，2），50个样本，二维特征，在平面上展示该样本点
    ax.scatter(X2[:, 0], X2[:, 1], color='r', label="Class -1")
    
    # 设置坐标轴范围
    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)
    ax.set_xlabel("Feature 1")
    ax.set_ylabel("Feature 2")
    ax.legend()
    
    # 计算决策边界: w0 + w1*x + w2*x2 = 0  => x2 = -(w0 + w1*x) / w2 现在的图，是纵轴是x2，横轴是x1
    if w[2] != 0:  # 避免 w2=0 除零错误
        x_vals = np.array([-4, 4])#边界点，因为图的大小就是 -4  到  4，横坐标4对应的纵坐标是多少，-4对应的纵坐标是多少，然后两点连线
        y_vals = -(w[0] + w[1] * x_vals) / w[2]
        ax.plot(x_vals, y_vals, 'k-', lw=2)

    # 保存帧
    frame_path = f"{frame_folder}/frame_{frame:03d}.png"
    frame_paths.append(frame_path)
    plt.savefig(frame_path)#保存当前的图片
plt.close()

# 生成 GIF 动画
gif_path = "perceptron_training.gif"
imageio.mimsave(gif_path, [imageio.imread(frame) for frame in frame_paths], fps=2)#多张图片生成gif

print(f"GIF 动画已保存: {gif_path}")
```
